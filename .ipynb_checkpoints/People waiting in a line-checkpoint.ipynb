```python
import numpy as np

array1=np.arange(1,1001,1) # 1000 people waiting in a line
while len(array1)>1:
    del_index=np.arange(1,len(array1),2) # Remove people in even places recursively
    new=np.delete(array1,del_index)
    array1=new
    print(array1)
```

```python
print('The person left in the queue will be the one standing at the front of the initial queue')
```
